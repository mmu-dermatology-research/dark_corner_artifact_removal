#### *Inpainting Method Activations Comparison*

*For Script*
* The first GradCAM activation example shows comparisons between the classification of the baseline model against both inpainting method models.
* An image of a known 'Other' classification lesion has been used. On the left column we can see that it was incorrectly predicted as melanoma by the model, and that the heatmap is focused predominantly on the DCA region.
* This suggests that deep learning networks do infact use the DCA for prediction
* The results for both the Navier Stokes and Telea method model predictions both correctly identify the lesion as belonging to the other class and we can see that the activations focus more on the intended area and less on the outer area where the DCA once was.

```python
exp.gradcam_method_comparison(base_model, ns_model, telea_model)
```

#### *GradCAM Inscribed DCA*

*For Script*
* For the second GradCAM example, an image with the known class 'Other' and correct prediction with the baseline model is used and versions with superimposed DCA are created.
* GradCAM images are generated for each of these generated images

* As this happens, the classification changes to incorrectly identify the image as 'Melanoma'.
* This further supports the suggestion that deep learning networks are using DCA for classification predictions.
