## Retrieve all of the image contours

The next few steps mirror that done in the automated masking process. This time we are interested in the radius and centre coordinates though.

```python
# Find the contours on the image - again
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_RGB2GRAY)
image_thresh = 100
ret,thresh = cv2.threshold(test_image_gray, image_thresh, 255, cv2.THRESH_BINARY)
contours,hierarchy = cv2.findContours(thresh, 1, 2)
cnt = contours[0]

img_contours = np.zeros(test_image.shape)
#plt.figure(figsize = (6,6))
#plt.imshow((cv2.drawContours(img_contours, contours, -1, (0,255,0), 3)).astype(np.uint8));

# get the largest again..
contoursB = []
big_contour = []
max = 0
for i in contours:
    area = cv2.contourArea(i) #--- find the contour having biggest area ---
    if(area > max):
        max = area
        big_contour = i 
        contoursB.append(i)

# find the minimum enclosing circle centre coordinates and radius
(x,y), radius = cv2.minEnclosingCircle(big_contour)
center = (int(x), int(y))

# negate from the radius to account for the 'transition zone' into the DCA
radius = int(radius) - 2

# create the new masking circle using the centre point and the new radius
this_contour = cv2.circle(np.ones(test_image_gray.shape),center,radius,(0,255,0),-1)





# display the new mask
plt.figure(figsize = (12,8))
plt.subplot(1,2,1)
plt.imshow(this_contour, cmap = 'gray')

# identify the center of the circle
this_contour[center[1]][center[0]] = 1

plt.subplot(1,2,2)
plt.imshow(this_contour, cmap = 'gray');
this_contour[center[1]][center[0]] = 0
```

## Calculate the margins that are available for removal

```python
# calculate the edge differences..
# how much white space beyond the radius?
l_edge = center[0] - radius if center[0] - radius >= 0 else 0
r_edge = 224 - (center[0] + radius) if center[0] + radius < 224 else 0
t_edge = center[1] - radius if center[1] - radius >= 0 else 0
b_edge = 224 - (center[1] + radius) if center[1] + radius < 224 else 0


# calculate the vert and horiz totals
vertical = t_edge + b_edge
horizontal = l_edge + r_edge

# take the smallest value of the 2, this is the maximum we can remove to keep the image square
removal = min([vertical, horizontal])

print(center)
print("Left: ", l_edge)
print("Right: ", r_edge)
print("Top: ", t_edge)
print("Bottom: ", b_edge, "\n")
print("Horizontal: ", horizontal)
print("Vertical: ", vertical, "\n")
print("To remove: ", removal)


# calculate how much to actually remove from the image
# how many pixels left to remove?
vertical_removal = removal
horizontal_removal = removal

new_top = t_edge if t_edge <= vertical_removal else vertical_removal
vertical_removal -= new_top
new_bottom = 224 - vertical_removal

new_left = l_edge if l_edge <= horizontal_removal else horizontal_removal
horizontal_removal -= new_left
new_right = 224 - horizontal_removal

print(new_left, "pixels removed from the left. New coordinate:", new_left)
print(horizontal_removal, "pixels removed from the right. New coordinate:", new_right)
print(new_top, "pixels removed from the left. New coordinate:", new_top)
print(vertical_removal, "pixels removed from the right. New coordinate:", new_bottom)

cropped_mask = np.copy(this_contour[new_top:new_bottom, new_left:new_right])
cropped_image = np.copy(test_image[new_top:new_bottom, new_left:new_right])
```
