```python
lesions_train = X[0]
lesions_test = X[1]

#testing
lesions_train_test = Y[0]
lesions_test_test = Y[1]


lesions_train_copy = np.copy(lesions_train.images)
lesions_train_copy_test = np.copy(lesions_train_test.images)
```

```python
image = lesions_train.images[2]
image_test = lesions_train_test.images[2]
#plt.imshow(image, cmap = 'viridis')
fig, axes = plt.subplots(1,2)


axes[0].imshow(image_test, cmap = 'gray')
axes[1].imshow(image, cmap = 'viridis')

fig.suptitle('original images');
```

```python
test_gamma_mod = im.modify_gamma(image_test,1)

plt.figure(figsize=(5,5))
plt.title('Gamma corrected image')
plt.imshow(test_gamma_mod, cmap = 'gray')
test_gamma_mod.shape
```

```python
#Gamma Encoding
gamma = 0.1  #Gamma < 1 ~ Dark ; Gamma > 1 ~ Bright

gamma_correction = ((image_test/255) ** (1/gamma))

plt.figure(figsize=(20,20))
plt.title('Gamma corrected image')
plt.imshow(gamma_correction, cmap = 'gray')
gamma_correction.shape
```

```python
mask_gam = gamma_correction > 0.4
gamma_correction[mask_gam] = 0

plt.imshow(gamma_correction, cmap = 'gray')
```
